```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from cifar10 import CIFAR10
```

```python
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'boat', 'truck']

random_indices = []
for idx in range(10):
    random_indices += list(np.random.choice(np.where(cifar.dev.data['labels']==idx)[0], size=10))

fig = plt.figure(figsize=(15,15))
for i, ridx in enumerate(random_indices):
    ax = fig.add_subplot(10,10,i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    if not i % 10:
        ax.set_ylabel(classes[i//10])
    
    tmp_picture = cifar.dev.data['images'][ridx,...]
    ax.imshow(tmp_picture)
    
plt.show()
```
